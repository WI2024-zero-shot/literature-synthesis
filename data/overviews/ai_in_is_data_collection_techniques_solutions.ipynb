```python
import pandas as pd
import glob
from unstructured.partition.pdf import partition_pdf
```

```python
pd.set_option('display.max_colwidth', None)
pd.set_option("display.max_rows", None)
```

```python

# Specify the path to your CSV file
csv_file_path = 'ai_in_is_data_collection_techniques_solutions.csv'

# Use pandas to read the CSV file
df = pd.read_csv(csv_file_path, sep=";", encoding='cp1252')
```

```python
df.head(1)
```

```python
my_dict = {}
x = glob.glob("*.pdf")
for elem in x:
    z = str(elem)
    elements = partition_pdf(z)
    joined_elements = "\n\n".join([str(el) for el in elements])
    my_dict[z] = joined_elements
    # Returns a List[Element] present in the pages of the parsed pdf document

```

```python
df_new = pd.Series(my_dict)
df_new.index = df_new.index.str.replace('.pdf', '')
df_new2 = df_new.reset_index()
df_new2.columns = ['u', 'text']
df_new2['u'] = df_new2['u'].astype(str)
df_new2['text'] = df_new2['text'].str.replace('\n\n', ' ')

```

```python
merged_df = pd.merge(df, df_new2, left_index=True, right_index=True, how='inner')
merged_df = merged_df.drop('u', axis=1)
```
